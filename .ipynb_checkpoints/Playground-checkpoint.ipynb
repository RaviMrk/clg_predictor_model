```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
df = pd.read_csv('pict_comp.csv')
df.head()
```

```python
cat = ['LSCO', 'LSCH', 'LOPENO', 'LOPENH', 'LOBCO',
       'LOBCH', 'GVJO', 'GVJH', 'GSTO', 'GSTH', 'GSCO', 'GSCH', 'GOPENO',
       'GOPENH', 'GOBCO', 'GOBCH', 'GNT3O', 'GNT3H', 'GNT2O', 'GNT2H', 'GNT1O',
       'GNT1H']
```

```python
from statsmodels.tsa.ar_model import AR
```

```python
for i in cat:
    df_t = df[['Year', i]]
    range_date = pd.date_range(start ='2013', end ='2019', freq='A')
    df_t.plot(x='Year', y=i, color='red')
    print(i)
    plt.show()
    df_t1 = df_t[i]
    model = AR(df_t1)
    model_fit = model.fit(1)
    yhat = model_fit.predict(len(df_t1), len(df_t1))
    yhat = yhat.astype(int)
    print(yhat)
```
